```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataframe = pd.read_csv("affirmativeaction_ucsc_ucsb_ucsd.csv")

dataframe
```

```python
UCSCdata = dataframe[dataframe["College"] == "UCSC"]
```

```python
AmericanIndianAdmitted = list(UCSCdata["American Indian Admitted"])
AmericanIndianApplied = list(UCSCdata["American Indian Applied"])
HispanicAdmitted = list(UCSCdata["Hispanic/Latinx Admitted"])
HispanicApplied = list(UCSCdata["Hispanic/Latinx Applied"])
AfricanAmericanAdmitted = list(UCSCdata["African American Admitted"])
AfricanAmericanApplied = list(UCSCdata["African American Applied"])
AsianAdmitted = list(UCSCdata["Asian/Pacific Islander Admitted"])
AsianApplied = list(UCSCdata["Asian/Pacific Islander Applied"])
WhiteAdmitted = list(UCSCdata["White Admitted"])
WhiteApplied = list(UCSCdata["White Applied"])
MaleAdmitted = list(UCSCdata["Male Admitted"])
MaleApplied = list(UCSCdata["Male Applied"])
FemaleAdmitted = list(UCSCdata["Female Admitted"])
FemaleApplied = list(UCSCdata["Female Applied"])


# to calculate average GPAs for a given year
FirstGPAAdmitted = list(UCSCdata["4.20 and Above Admitted"])
FirstGPAApplied = list(UCSCdata["4.20 and Above Applied"])
SecondGPAAdmitted = list(UCSCdata["3.80 - 4.19 Admitted"])
SecondGPAApplied = list(UCSCdata["3.80 - 4.19 Applied"])
ThirdGPAAdmitted = list(UCSCdata["3.40 - 3.79 Admitted"])
ThirdGPAApplied = list(UCSCdata["3.40 - 3.79 Applied"])
FourthGPAAdmitted = list(UCSCdata["3.00 - 3.39 Admitted"])
FourthGPAApplied = list(UCSCdata["3.00 - 3.39 Applied"])
FifthGPAAdmitted = list(UCSCdata["0.00 - 2.99 Admitted"])
FifthGPAApplied = list(UCSCdata["0.00 - 2.99 Applied"])

# reversing ethnicities admittance 
AmericanIndianAdmitted = AmericanIndianAdmitted[::-1]
AmericanIndianApplied = AmericanIndianApplied[::-1]
HispanicAdmitted = HispanicAdmitted[::-1]
HispanicApplied = HispanicApplied[::-1]
AfricanAmericanAdmitted = AfricanAmericanAdmitted[::-1]
AfricanAmericanApplied = AfricanAmericanApplied[::-1]
AsianAdmitted = AsianAdmitted[::-1]
AsianApplied = AsianApplied[::-1]
WhiteAdmitted = WhiteAdmitted[::-1]
WhiteApplied = WhiteApplied[::-1]

MaleAdmitted = MaleAdmitted[::-1]
MaleApplied = MaleApplied[::-1]
FemaleAdmitted = FemaleAdmitted[::-1]
FemaleApplied = FemaleApplied[::-1]


# reversing GPA
FirstGPAAdmittedReversed = FirstGPAAdmitted[::-1]
FirstGPAAppliedReversed = FirstGPAApplied[::-1]
SecondGPAAdmittedReversed = SecondGPAAdmitted[::-1]
SecondGPAAppliedReversed = SecondGPAApplied[::-1]
ThirdGPAAdmittedReversed = ThirdGPAAdmitted[::-1]
ThirdGPAAppliedReversed = ThirdGPAApplied[::-1]
FourthGPAAdmittedReversed = FourthGPAAdmitted[::-1]
FourthGPAAppliedReversed = FourthGPAApplied[::-1]
FifthGPAAdmittedReversed = FifthGPAAdmitted[::-1]
FifthGPAAppliedReversed = FifthGPAApplied[::-1]


# percentage arrays for ethnicities
percentAmericanIndianAdmitted = []
percentHispanicAdmitted = []
percentAfricanAmericanAdmitted = []
percentAsianAdmitted = []
percentWhiteAdmitted = []
percentMaleAdmitted = []
percentFemaleAdmitted = []

firstGPAPercentage = []
secondGPAPercentage = []
thirdGPAPercentage = []
fourthGPAPercentage = []
fifthGPAPercentage = []

# average GPAs 
averageGPA = []

for i in range(0, 20, 1):
    total = FirstGPAAppliedReversed[i] + SecondGPAAppliedReversed[i] + ThirdGPAAppliedReversed[i] + FourthGPAAppliedReversed[i] + FifthGPAAppliedReversed[i]
    average = (FirstGPAAppliedReversed[i] * 4.2 + SecondGPAAppliedReversed[i] * 4.0 + ThirdGPAAppliedReversed[i] * 3.6 + FourthGPAAppliedReversed[i] * 3.2 + FifthGPAAppliedReversed[i] * 3.0) / total 
    averageGPA.append(average)
    
    AmericanIndianPercentage = AmericanIndianAdmitted[i] / AmericanIndianApplied[i]
    HispanicAdmittedPercentage = HispanicAdmitted[i] / HispanicApplied[i]
    AfricanAmericanPercentage = AfricanAmericanAdmitted[i] / AfricanAmericanApplied[i]
    AsianPercentage = AsianAdmitted[i] / AsianApplied[i]
    WhitePercentage = WhiteAdmitted[i] / WhiteApplied[i]
    
    MalePercentage = MaleAdmitted[i] / MaleApplied[i]
    FemalePercentage = FemaleAdmitted[i] / FemaleApplied[i]
    
    percentAmericanIndianAdmitted.append(AmericanIndianPercentage)
    percentHispanicAdmitted.append(HispanicAdmittedPercentage)
    percentAfricanAmericanAdmitted.append(AfricanAmericanPercentage)
    percentAsianAdmitted.append(AsianPercentage)
    percentWhiteAdmitted.append(WhitePercentage)
    percentMaleAdmitted.append(MalePercentage)
    percentFemaleAdmitted.append(FemalePercentage)
    
    firstGPAPercentage.append(FirstGPAAdmittedReversed[i]/FirstGPAAppliedReversed[i])
    secondGPAPercentage.append(SecondGPAAdmittedReversed[i]/SecondGPAAppliedReversed[i])
    thirdGPAPercentage.append(ThirdGPAAdmittedReversed[i]/ThirdGPAAppliedReversed[i])
    fourthGPAPercentage.append(FourthGPAAdmittedReversed[i]/FourthGPAAppliedReversed[i])
    fifthGPAPercentage.append(FifthGPAAdmittedReversed[i]/FifthGPAAppliedReversed[i])

    

years = []
numYear = 2000
for i in range(0, 20, 1):
    years.append(numYear)
    numYear += 1

# Percentage Arrays of Ethnicities
percentageAmericanAdmittedArray = np.array(percentAmericanIndianAdmitted)
percentageHispanicAdmittedArray = np.array(percentHispanicAdmitted)
percentageAfricanAmericanAdmittedArray = np.array(percentAfricanAmericanAdmitted)
percentageAsianAdmittedArray = np.array(percentAsianAdmitted)
percentageWhiteAdmittedArray = np.array(percentWhiteAdmitted)

percentageMaleAdmittedArray = np.array(percentMaleAdmitted)
percentageFemaleAdmittedArray = np.array(percentFemaleAdmitted)


# just for GPAs
FirstAmericanAdmittedArray = np.array(firstGPAPercentage)
SecondAmericanAdmittedArray = np.array(secondGPAPercentage)
ThirdAmericanAdmittedArray = np.array(thirdGPAPercentage)
FourthAmericanAdmittedArray = np.array(fourthGPAPercentage)
FifthAmericanAdmittedArray = np.array(fifthGPAPercentage)

AverageGPAArray = np.array(averageGPA)

    

    
for i in range(0, AverageGPAArray.size, 2):
    print(years[i], " to: ", years[i+1])
    
    
    differenceAdmittedAmericanIndian = percentageAmericanAdmittedArray[i+1] - percentageAmericanAdmittedArray[i]
    differenceAdmittedHispanic = percentageHispanicAdmittedArray[i+1] - percentageHispanicAdmittedArray[i]
    differenceAdmittedAfricanAmerican = percentageAfricanAmericanAdmittedArray[i+1] - percentageAfricanAmericanAdmittedArray[i]
    differenceAsian = percentageAsianAdmittedArray[i+1] - percentageAsianAdmittedArray[i]
    differenceWhite = percentageWhiteAdmittedArray[i+1] - percentageWhiteAdmittedArray[i]
    
    differenceMale = percentageMaleAdmittedArray[i+1] - percentageMaleAdmittedArray[i]
    differenceFemale = percentageFemaleAdmittedArray[i+1] - percentageFemaleAdmittedArray[i]

    
    
    differenceGPA = AverageGPAArray[i+1]/4.2 - AverageGPAArray[i]/4.2
#    print("Comparing: ", AverageGPAArray[i+1], "to: ", AverageGPAArray[i], " difference: ", differenceGPA)
    roundedAdmittedAmericanIndian = int(differenceAdmittedAmericanIndian * 1000) / 1000 * 100
    roundedAdmittedHispanic = int(differenceAdmittedHispanic * 1000) / 1000 * 100
    roundedAdmittedAfricanAmerican = int(differenceAdmittedAfricanAmerican * 1000) / 1000 * 100 
    roundedAdmittedAsian = int(differenceAsian * 1000) / 1000 * 100
    roundedAdmittedWhite = int(differenceWhite * 1000) / 1000 * 100
    roundedAdmittedMale = int(differenceMale * 1000) / 1000 * 100
    roundedAdmittedFemale = int(differenceFemale * 1000) / 1000 * 100
    
    
    
    roundedGPA = int(differenceGPA * 1000) / 1000 * 100
    print("GPA increase/decrease: " + str(roundedGPA))
    print ("Admitted American Indian: " + str(round(roundedAdmittedAmericanIndian, 2)) + "%")
    print ("Admitted Hispanic: " + str(round(roundedAdmittedHispanic, 2)) + "%")
    print ("Admitted African American: " + str(round(roundedAdmittedAfricanAmerican, 2)) + "%")
    print ("Admitted Asian: " + str(round(roundedAdmittedAsian, 2)) + "%")
    print ("Admitted White: " + str(round(roundedAdmittedMale, 2)) + "%")
    print ("Admitted Female: " + str(round(roundedAdmittedFemale, 2)) + "%")
    print ("Admitted Male: " + str(round(roundedAdmittedMale, 2)) + "%")


    

    print("")
    
print("size: ", AverageGPAArray.size)
#plt.plot(years, FirstAmericanAdmittedArray, color='green', label='4.20 and Above Admitted')
#plt.plot(years, SecondAmericanAdmittedArray, color='red', label='3.80 - 4.19 Admitted')
#plt.plot(years, ThirdAmericanAdmittedArray, color='pink', label='3.40 - 3.79 Admitted')
#plt.plot(years, FourthAmericanAdmittedArray, color='purple', label='3.00 - 3.39 Admitted')
#plt.plot(years, FifthAmericanAdmittedArray, color='orange', label='0.00 - 2.99 Admitted')


```
